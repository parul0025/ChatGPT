```python
# Define a function to generate a chatbot response
def generate_chat_response(message, conversation):
    conversation.append({'role': 'user', 'content': message})

    response = openai.Completion.create(
        model='gpt-3.5-turbo',
        prompt=conversation,
    )

    # Retrieve and return the chatbot's reply
    reply = response.choices[0].message.content
    return reply
```

```python
# Set up initial conversation
initial_conversation = [
    {'role': 'system', 'content': 'You are a helpful assistant.'},
    {'role': 'user', 'content': 'Hello, how are you?'}
]

# Chat loop
while True:
    user_input = input("User: ")

    # Generate chatbot response
    bot_response = generate_chat_response(user_input, initial_conversation)

    print("ChatBot:", bot_response)

    # Update conversation with new messages
    initial_conversation.append({'role': 'user', 'content': user_input})
    initial_conversation.append({'role': 'assistant', 'content': bot_response})
```
